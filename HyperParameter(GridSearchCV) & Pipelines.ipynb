```python
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
import seaborn as sns
```

```python
import pandas as pd
df=pd.read_csv("hd1.csv")
df.head()
```

```python
X = df[['age', 'sex', 'cp', 'chol', 'restecg', 'ca', 'thal', 'slope']]
y = df['target']
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
## Pieplining
numeric_preprocessor = Pipeline(
    steps=[
        ("imputation_mean", SimpleImputer(missing_values=np.nan, strategy="mean")),
        ("scaler", StandardScaler()),
    ]
)
```

```python
from sklearn import set_config
```

```python
set_config(display='diagram')
```

```python
numeric_preprocessor
```

```python
categorical_preprocessor = Pipeline(
    steps=[
        (
            "imputation_constant",
            SimpleImputer(fill_value="missing", strategy="constant"),
        ),
        ("onehot", OneHotEncoder(handle_unknown="ignore")),
    ]
)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer([
    ("numerical",numeric_preprocessor,["age","cp"])]


)

# preprocessor=ColumnTransformer(
#     [("categorical",categorical_preprocessor,["sex"]),
#     ("numerical",numeric_preprocessor,["age","cp"])]


# )   #here we dont want this 
```

```python
#2 ways for creating pipeline
#1way
# from sklearn.pipeline import make_pipeline
# pipe1=make_pipeline(preprocessor,RandomForestRegressor())
#2way
steps=[("preprocessor",preprocessor),("Model",RandomForestRegressor())]
pipe1=Pipeline(steps)
```

```python
pipe1
```

```python
pipe1.fit(X_train,y_train)
```

```python
y_pred=pipe1.predict(X_test)
```

```python
import numpy as np
y_pred=np.round(y_pred)
y_pred
```

```python
#hyperparameter tuning
param_grid={
    "Model__n_estimators":[1,10,20],
    "Model__max_features":["auto"],
}
```

```python
grid_search=GridSearchCV(pipe1,param_grid=param_grid)
grid_search
```

```python
grid_search.fit(X_train,y_train)
```

```python
y_pred=grid_search.predict(X_test)
```

```python
y_pred
```

```python
y_pred=np.round(y_pred)
y_pred
```

```python
grid_search.best_params_
```
